```python
c = tf.constant([-40, -10, 40 , 0, 15, 22, 30, 38, 50, 80], dtype=tf.float32)
f = tf.constant([-40, 14, 104, 32, 59, 72, 86, 100, 122, 176], dtype=tf.float32)

c,f
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

model.fit(c,f, epochs=1000, verbose=False)
print("finished")
```

```python
print(model.predict([12]))
```

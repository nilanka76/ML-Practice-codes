```python
import csv

header = ['Name', 'Happiness Score', 'Health Score', 'State']
data = [
    ['A', 7, 7, 'HAPPY'],
    ['B', 2, 1, 'SAD'],
    ['C', 3, 2, 'SAD'],
    ['D', 4, 5, 'NEUTRAL'],
    ['E', 5, 4, 'NEUTRAL'],
    ['F', 2, 2, 'SAD'],
    ['G', 9, 8, 'HAPPY'],
    ['H', 3, 2, 'SAD'],
    ['I', 4, 5, 'NEUTRAL'],
    ['J', 5, 4, 'NEUTRAL'],
    ['K', 3, 2, 'SAD'],
    ['L', 2, 3, 'SAD'],
    ['M', 3, 1, 'SAD'],
    ['N', 7, 6, 'HAPPY'],
    ['O', 5, 5, 'NEUTRAL'],
    ['P', 8, 9, 'HAPPY'],

]

with open('happiness.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)

    writer.writerow(header)

    writer.writerows(data)
```

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("happiness.csv")
dataset
```

```python
x = dataset.drop(columns=["Name", "State"])
x
```

```python
y = dataset["State"]
y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import datasets, metrics

classifier = GaussianNB()
classifier.fit(x, y)
score = metrics.accuracy_score(y, classifier.predict(x)) # I use the same data as train data which I should not, but had to, because I made less data due to sortage of time
print("Accuracy: %f" % score)
```
